```python
import pandas as pd
import numpy as np
import os


from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
%matplotlib inline
```

```python
start_dir = os.getcwd()
os.chdir("..")
path_to_data = os.getcwd() + os.path.join(os.path.sep, 'data')
path_to_clean_stats = os.getcwd() + os.path.join(os.path.sep, 'data', 'clean_stats.tsv')
os.chdir(start_dir)
print(path_to_clean_stats)
```

```python
result = pd.read_csv(path_to_clean_stats, sep='\t')
result["pred_perf"] = ""
result.head()
```

```python
for index, row in result.iterrows():
    row.pred_perf = np.mean(eval(row.perf))
result.head()
```

```python
final = result[['player','pos','salary','pred_perf']]
final.set_index('player')
```

```python
path_to_final = os.path.join(path_to_data, 'final.csv')
final.to_csv(path_to_final,index=False)
```
